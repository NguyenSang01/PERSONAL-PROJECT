```python
#Tìm những hàng có giá trị bất thường trong data
mask = df.applymap(lambda cell: ('Test' in str(cell)) or ('TEST' in str(cell)) or ('?????' in str(cell))).any(axis=1)

# Xóa các hàng dựa trên mặt nạ
df = df[~mask]
```

```python
# Tìm những giá trị bị trùng trong cột "EmployeeNumber"

duplicates = df[df.duplicated(subset='EmployeeNumber', keep=False)]

# Xóa các hàng bị trùng
df = df.drop_duplicates(subset='EmployeeNumber', keep='first')
```

```python
# Chọn các hàng thỏa mãn điều kiện Education <= 3 và Age - TotalWorkingYears <= 18
condition1 = (df['Education'] <= 3) & (df['Age'] - df['TotalWorkingYears'] <= 18)

# Chọn các hàng thỏa mãn điều kiện Education > 3 và Age - TotalWorkingYears <= 22
condition2 = (df['Education'] > 3) & (df['Age'] - df['TotalWorkingYears'] <= 22)

# Cập nhật giá trị của cột TotalWorkingYears
df.loc[condition1, 'TotalWorkingYears'] = df.loc[condition1, 'Age'] - 18
df.loc[condition2, 'TotalWorkingYears'] = df.loc[condition2, 'Age'] - 22
```

```python
grouped_df_1 = df.groupby(['Education', 'Age','TotalWorkingYears'])['EmployeeNumber'].count()
grouped_df_1
```

```python
df['Attrition']=df['Attrition'].apply(lambda x: 1 if x=='Voluntary Resignation' else 0)
```

```python
bins = [18, 20, 30, 40, 50, float('inf')]
labels = ['18-20', '21-30', '31-40', '41-50', 'Over 50']

# Tạo cột mới "Age_group" chứa nhóm tuổi tương ứng
df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df
```

```python
# Định nghĩa khoảng và nhãn tương ứng cho các nhóm khoảng cách
df['DistanceFromHome'] = df['DistanceFromHome'].astype(float)
bins = [0, 5, 10, 20, float('inf')]
labels = ['0-5km', '5-10km', '10-20km', 'Over 20km']

# Tạo cột mới "Distance_Group" chứa nhóm khoảng cách tương ứng
df['Distance_Group'] = pd.cut(df['DistanceFromHome'], bins=bins, labels=labels, right=False)
df['Distance_Group'].unique()
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].astype(float)
bins = [0, 5000, 10000, 15000, float('inf')]
labels = ['0-5.000', '5.000-10.000', '10.000-15.000','Over 15.000']

# Tạo cột mới "Age_group" chứa nhóm tuổi tương ứng
df['MonthlyIncome_group'] = pd.cut(df['MonthlyIncome'], bins=bins, labels=labels, right=False)
```

```python

bins = [0, 5, 10, 15, float('inf')]
labels = ['0-5 year', '5-10 year','10-15 year',' Over 15 year']

# Tạo cột mới "Age_group" chứa nhóm tuổi tương ứng
df['Years_group'] = pd.cut(df['YearsAtCompany'], bins=bins, labels=labels, right=False)
df
```

```python
# Kiểm tra thông tin JobRole và Depart ment
grouped_df = df.groupby(['Department', 'JobRole'])['EmployeeNumber'].count()
grouped_df
```

```python
# Gán giá trị cho cột 'Department' dựa trên giá trị của cột 'JobRole'
df.loc[df['JobRole'] == 'Human Resources', 'Department'] = 'Human Resources'
df.loc[df['JobRole'].isin(['Sales Executive', 'Sales Representative',
'Healthcare Representative']), 'Department'] = 'Sales'
df.loc[df['JobRole'].isin(['Manufacturing Director', 'Research Director',
'Research Scientist', 'Laboratory Technician']), 'Department'] = 'Research & Development'
```

```python
grouped_df = df.groupby(['Department', 'JobRole'])['EmployeeNumber'].count()
grouped_df
```

```python
import matplotlib.pyplot as plt

# Chia cột 'Age' thành 10 phần
df['Age_bins'] = pd.cut(df['Age'], bins=10)

# Tính tỉ lệ nghỉ việc (Attrition) theo giá trị 'Yes' chia cho cột 'EmployeeNumber'
attrition_rate = df.loc[df['Attrition'] == 'Yes'].groupby('Age_bins')['EmployeeNumber'].count() / df.groupby('Age_bins')['EmployeeNumber'].count()

# Vẽ biểu đồ histogram
plt.barh(attrition_rate.index.astype(str), attrition_rate.values)
plt.xlabel('Attrition Rate')
plt.ylabel('Age Bins')
plt.title('Attrition Rate by Age Bins')

# Làm tròn giá trị cột 'Age' trên trục y
plt.yticks(range(len(attrition_rate.index)), [str(round(age_bin.left)) + '-' + str(round(age_bin.right)) for age_bin in attrition_rate.index])

plt.show()
```

```python
df1 = df[df['Attrition'] == 'Yes']
df1
```

```python
monthly_income_q25 = df1['MonthlyIncome'].quantile(0.25)
monthly_income_q50 = df1['MonthlyIncome'].quantile(0.5)
monthly_income_q75 = df1['MonthlyIncome'].quantile(0.75)
print("Tứ phân vị 25:", monthly_income_q25)
print("Tứ phân vị 50:", monthly_income_q50)
print("Tứ phân vị 75:", monthly_income_q75)
```

```python
plt.hist(df1['MonthlyIncome'],rwidth=0.8)
plt.xlabel('Monthly income')
plt.ylabel('Tần số')
plt.title('Đồ thị phân bố Monthly income')
plt.grid(True)
plt.axvline(x=monthly_income_q25, color='r', linestyle='--',label = '0.25 Deciles')
plt.text(monthly_income_q25, 1250, format(monthly_income_q25, ","), ha='left',  color='red')

plt.axvline(x=monthly_income_q50, color='g', linestyle='--',label = '0.5 Deciles')
plt.text(monthly_income_q50, 1300, format(monthly_income_q50,","), ha='left',  color='red')

plt.axvline(x=monthly_income_q75, color='b', linestyle='--',label = '0.75 Deciles')
plt.text(monthly_income_q75, 1300, format(monthly_income_q75, ","), ha='left',color='red')
plt.legend()
```
